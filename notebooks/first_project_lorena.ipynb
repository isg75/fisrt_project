```python
patents_mapping = {'Grants of patents (number)': 1, 'Resident patent filings (per million population)': 2, "Patents in force (number)":3} 
data['patents_mapped'] = data['patents'].map(patents_mapping)
display(data)
```

```python
data['patents_mapped'].value_counts()
# Number of granted patents: 771
# Patent filed per resident: 631
# Number of current and valid patents: 458
```

```python
condition1 = data.patents_mapped ==1
condition1
```

```python
filtered_data = data[condition1]
filtered_data
```

```python
data['year'] = pd.to_numeric(data['year'], errors='coerce')
```

```python
data['number_patents'] = pd.to_numeric(data['number_patents'], errors='coerce')
```
